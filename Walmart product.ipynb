```python
# 8.Calculate the monthly sales and find out the trends
```

```python
data['transaction_date']=pd.to_datetime(data['transaction_date'])
```

```python
data
```

```python
data['month']=data['transaction_date'].dt.month
```

```python
data
```

```python
data['month']=data['transaction_date'].dt.month_name()
```

```python
data
```

```python
monthly_revenue=data.groupby('month')['revenue'].sum().reset_index()
```

```python
monthly_revenue
```

```python
monthly_revenue=data.groupby('month')['revenue'].sum().reset_index()
plt.figure(figsize=(12,6))
plt.plot(monthly_revenue['month'],monthly_revenue['revenue'],marker='o',linestyle='-',ms='10',mfc='r',color='c')
plt.title('monthly revenue trend')
plt.xlabel('month')
plt.ylabel('revenue')
plt.grid()
plt.show()
```
